```python
dfs = []
for file in ['PL1.csv', 'PL2.csv', 'PL3.csv']:
    df = pd.read_csv(file, header=None)
    df['mean'] = df.mean(axis=1)
    dfs.append(df)
```

```python
df_ctr = 0
fig, ax = plt.subplots(figsize = (18,5), linewidth = 1, nrows=3, ncols=1)
for ax_t in ax:
    ax_t.plot(dfs[df_ctr]['mean'])
    ax_t.set_title(f'PL{df_ctr+1} mean: {round(dfs[df_ctr].iloc[:,-1].sum() / 60 / 1000, 2)}kWh p.a.')
    df_ctr += 1
plt.tight_layout()
plt.show()
```
